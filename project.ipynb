```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Corona']=le.fit_transform(df['Corona'])
df['Cough_symptoms']=le.fit_transform(df['Cough_symptoms'])
df['Fever']=le.fit_transform(df['Fever'])
df['Sore_throat']=le.fit_transform(df['Sore_throat'])
df['Shortness_of_breath']=le.fit_transform(df['Shortness_of_breath'])
df['Headache']=le.fit_transform(df['Headache'])
df['Age_60_above']=le.fit_transform(df['Age_60_above'])
df['Sex']=le.fit_transform(df['Sex'])
```

```python
df.info()
```

```python
df
```

```python
 col = list(df.columns)
 for i in col:
    if df[i].dtypes== 'int64'or df[i].dtypes=='float64':
        sns.boxplot(df[i])
        plt.show()
```

```python
q1 = df.Cough_symptoms.quantile(0.25)
q3 = df.Cough_symptoms.quantile(0.75)
iqr = q3-q1
data = df[(df.Cough_symptoms >= q1 - 1.5*iqr) & (df.Cough_symptoms <= q3 + 1.5*iqr)]

q1 = df.Fever.quantile(0.25)
q3 = df.Fever.quantile(0.75)
iqr = q3-q1
df = df[(df.Fever >=q1-1.5*iqr) & (df.Fever <=q3+1.5*iqr)]
```

```python
for i in col:
    if data[i].dtypes== 'int64'or data[i].dtypes=='float64':
        sns.boxplot(data[i])
        plt.show()
```

```python
df.isnull().sum()
```

```python
df_new=df.drop(columns=['Age_60_above','Sex'])
df_new
```

```python
X = df.drop(columns=['Ind_ID', 'Test_date', 'Corona'])
y=df['Corona']
X = pd.get_dummies(X, drop_first=True)
```

```python
!pip install imblearn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)
```

#Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(criterion='entropy', random_state=0)
dt_model.fit(X_train, y_train)


```

```python
y_dt_pred= dt_model.predict(X_test)
y_dt_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score

dt_cr=classification_report(y_test, y_dt_pred)
print(dt_cr)
```

#Random forest

```python

from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
RandomForestClassifier()
```

```python
y_rf_pred= rf_model.predict(X_test)
y_rf_pred
```

```python
rf_cr=classification_report(y_test, y_rf_pred)
print(rf_cr)
```

# K Nearest Neighbor


```python
from sklearn.neighbors import KNeighborsClassifier
knn_model= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )
knn_model.fit(X_train, y_train)
KNeighborsClassifier()
```

```python
y_knn_pred= knn_model.predict(X_test)
y_knn_pred
```

```python
gnb_cr=classification_report(y_test, y_knn_pred)
print(gnb_cr)
```
